```python
features, targets = datasets.load_breast_cancer(
    return_X_y=True,
    as_frame=True
)
```

### Train-test split

```python
prng = np.random.RandomState(42)
train_features, test_features, train_target, test_target = (
    model_selection.train_test_split(
        features,
        targets,
        random_state=prng,
        stratify=targets,
        test_size=0.1
    )
)
```

```python
feature_preprocessor = pipeline.make_pipeline(
    preprocessing.StandardScaler()
).set_output(transform="pandas")
```

### Using SGDClassifier

```python
ml_pipeline = pipeline.make_pipeline(
    feature_preprocessor,
    linear_model.SGDClassifier(),
    verbose=True,
)
_ = ml_pipeline.fit(train_features, train_target)
```

### Assess Performance

```python
train_predictions = ml_pipeline.predict(train_features)
report = metrics.classification_report(
    train_target,
    train_predictions
)
print(report)
```

```python
test_predictions = ml_pipeline.predict(test_features)
report = metrics.classification_report(
    test_target,
    test_predictions
)
print(report)
```
